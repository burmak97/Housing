```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
%matplotlib inline 
#%matplotlib inline to show plots in notebook
```

**Reading a file**

```python
housing = pd.read_csv('housing.csv')
```

**Visualisation of data in the dataset**

```python
sns.set_style('darkgrid')
housing.hist(bins = 40, figsize = (20,15))
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15, 15))
plt.title('California housing prices')
sns.scatterplot(x = housing.longitude,
                y = housing.latitude,
                size = housing.population,
                hue = housing.median_house_value,
                alpha = 0.7,
                legend= "auto");
```

**housing DF attributes corelation with median_house_value**

```python
housing.corrwith(housing.median_house_value).sort_values(ascending=False)
```
